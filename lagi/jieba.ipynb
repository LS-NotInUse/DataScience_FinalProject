```python
import json
import jieba
import jieba.analyse
import re
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud


with open("post_2017-1-1_2017-5-1.txt", "r") as reader:
    all_posts = json.load(reader)
    
for post in all_posts:
    #print ("###############")
    post['message']=re.sub('^https?:\/\/.*[\r\n]*', '',post['message'], flags=re.MULTILINE)
    post['message']=re.sub('\｜(.*)', '',post['message'], flags=re.MULTILINE)
#     print (post['message'])

# seg_list = jieba.cut_for_search(str(all_posts[0]['message']))
# print("Default Mode: " + "/ ".join(seg_list))
# print ("------------------------------")

all_tags=[]
for post in all_posts:
    tags = jieba.analyse.extract_tags(post['message'],topK=30,withWeight=True)
    for tag, weight in tags:
        print(tag + "," + str(int(weight * 10000)))
        all_tags.append(tag)
        
        
      

stopwords = {}.fromkeys(["沒有","一個","什麼","那個","我們","他們"])

wc = WordCloud( font_path="NotoSerifCJKtc-Black.otf",#設置字體
               background_color="white", #背景顏色
               max_words = 2000 ,        #文字雲顯示最大詞數
               stopwords=stopwords)      #停用字詞
print (",".join(all_tags))
t = ",".join(all_tags)
wc.generate(t)
# 視覺化呈現
plt.imshow(wc)
plt.axis("off")
plt.figure(figsize=(20,12), dpi = 400)
plt.show()
```
